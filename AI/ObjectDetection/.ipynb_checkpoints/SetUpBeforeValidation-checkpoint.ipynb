# 1) labels

```python
import os
import json
```

```python
origin_path1 = "E:/DataSets/01.데이터/2.Validation/라벨링데이터/VL/Safety/Crane/YT/"
origin_path2 = "E:/DataSets/01.데이터/2.Validation/라벨링데이터/VL/Safety/Crane/TwistLock/"

target_path = "E:/DataSets/Harbor_Data/valid/labels/"
```

```python
os.mkdir(target_path)
```

```python
yt_list = os.listdir(origin_path1)
len(yt_list) # 9,129
```

```python
for index, path in enumerate(yt_list):
    with open(origin_path1 + path, 'r', encoding='utf-8') as f:
        json_data = json.load(f)
    
    txt_file = open(target_path + json_data['images']['identifier'] + ".txt", 'w')
    
    print(index, json_data['images']['identifier'] + ".txt")
    
    for obj in json_data['annotations']: 
        cName = obj['bbox']['classid']
        if 'CONTAINER' in cName:
            txt_file.write("0 ")
        elif 'YT' in cName:
            txt_file.write("1 ")
        else:
            txt_file.write("2 ")
        
        s_x = obj['bbox']['points'][0][0] / 1920
        s_y = obj['bbox']['points'][0][1] / 1080
        e_x = obj['bbox']['points'][2][0] / 1920
        e_y = obj['bbox']['points'][2][1] / 1080
        
        print(s_x, s_y, e_x, e_y)
        
        txt_file.write("%f %f %f %f\n" % (s_x, s_y, e_x, e_y))
        
    txt_file.close()
    print()
```

```python
tl_list = os.listdir(origin_path2)
len(tl_list) # 557
```

```python
for index, path in enumerate(tl_list):
    with open(origin_path2 + path, 'r', encoding='utf-8') as f:
        json_data = json.load(f)
    
    txt_file = open(target_path + json_data['images']['identifier'] + ".txt", 'w')
    
    print(index, json_data['images']['identifier'] + ".txt")
    
    for obj in json_data['annotations']: 
        cName = obj['bbox']['classid']
        if 'TWISTLOCK' in cName:
            txt_file.write("3 ")
        
        s_x = obj['bbox']['points'][0][0] / 1920
        s_y = obj['bbox']['points'][0][1] / 1080
        e_x = obj['bbox']['points'][2][0] / 1920
        e_y = obj['bbox']['points'][2][1] / 1080
        
        print(s_x, s_y, e_x, e_y)
        
        txt_file.write("%f %f %f %f\n" % (s_x, s_y, e_x, e_y))
        
    txt_file.close()
    print()
```
